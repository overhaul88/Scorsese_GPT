```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings, SimpleDirectoryReader, VectorStoreIndex
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.postprocessor import SimilarityPostprocessor
```

```python
Settings.embed_model = HuggingFaceEmbedding(model_name = "BAAI/bge-small-en-v1.5")
Settings.llm = None
Settings.chunk_size = 512
Settings.chunk_overlap = 40
```

```python
documents = SimpleDirectoryReader("docs").load_data()
```

```python
print(len(documents))
```

```python
#store vectors into vector database
index = VectorStoreIndex.from_documents(documents)
```

```python
#set number of docs to retrieve
top_k = 5

#configure retriever
retriever = VectorIndexRetriever(index = index, similarity_top_k = top_k)
```

```python
#assemble query engine
query_engine = RetrieverQueryEngine(
    retriever = retriever,
    node_postprocessors=[SimilarityPostprocessor(similarity_cutoff=0.5)]
)
```

```python
from urllib.parse import quote_from_bytes
query = "What is the role of Deckard in the movie?"
response = query_engine.query(query)

#query documents
context = "Context: \n"
for i in range(top_k):
  context = context + response.source_nodes[i].text + "\n\n"

print(context)
```

```python
# prompt (with context)
prompt_template_w_context = lambda context, comment: f"""[INST]\n
You are ScorseseGPT, a movie analyst with a vast knowledge of film history, genres, critical reception, and audience preferences.\n
User will provide you with context about a movie, including its title, script, director, cast, and any additional details.\n
Based on this information you need to generate insights about the movie.\n
Your response will be judged on quality of synthesized answer: Clarity, coherence, and fidelity to the original sources.\n\n

{context}
\n
Please respond to the following query in a clear consise manner with detailed description in context.\n

{comment}
[/INST]
"""
```

```python
comment = "What is the role of Deckard in the movie?"
prompt = prompt_template_w_context(context, comment)

print(prompt)
```

```python
def prompter(comment):
  #query documents
  response = query_engine.query(comment)

  context = "Context: \n"

  for i in range(top_k):
    context = context + response.source_nodes[i].text + "\n\n"

  prompt = prompt_template_w_context(context, comment)

  return prompt
```

```python
def scorsesegpt(input_query):
  input_prompt = prompter(input_query)

  inputs = tokenizer(input_prompt, return_tensors="pt")

  outputs = model.generate(input_ids=inputs["input_ids"].to("cuda"),
                           attention_mask=inputs["attention_mask"].to("cuda"),
                           pad_token_id=tokenizer.eos_token_id,
                           max_new_tokens=1024)

  result = tokenizer.batch_decode(outputs)[0]

  solution = result.split("[/INST]\n")[1].strip().replace("</s>", "\n~Scorsese-GPT")
  print(input_query + "\n" + solution)
```
